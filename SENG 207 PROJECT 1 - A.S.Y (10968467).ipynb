**_Note that there are null values in given dataset._**

### Removing the null values from the data set using the dropna() method

- **This method is going to get rid of all empty cells within the dataset.**
- **I'm employing this method because to make a good analysis, i would want to use data where all parameters are defined and so to achieve this i must take them out.**
* _There are other methods i could have employed here(for example,the fillna() method), which could be used to fill the null spaces with the mean/variance/standard deviation of that particular column where null cells are located. Now this particular method is not employed here because as a data analist, it would be more efficient to represent the statistical inferences/analysis at thier true values and not as approximates/speculating values - so that each conclusion is as precise as possible without any doubt or ambiguity._

```python
df.dropna(inplace= True)
```

```python
# Checking if there are any duplicates in the dataset.
df.duplicated()
```

```python
# Verifying that there are no more null cells.
df.isnull().any().any()
```

```python
# Showing the UPDATED dataset of all inventories of red wine.
df
```

#### 3. Showing statistical inferences of individual columns in the dataset.

### (i) fixed_acidity

```python
df.fixed_acidity.describe()
```

```python
mean_of_fixed_acidity = df.fixed_acidity.mean()
x = df.fixed_acidity < mean_of_fixed_acidity
x.value_counts()

# Finding how many wines have a fixed_acidity value less than the mean value of the fixed_acidity.
```

```python
lb = ['Below mean value','Above mean value']
explode = [0,0.1]
color_values = ['b', 'r']
plt.pie(x.value_counts(), labels= lb, autopct = '%.2f%%', explode= explode, colors= color_values)
plt.title('Percentage of wines with fixed_acidity above and below mean value')
plt.legend(x.value_counts())
```

```python
sns.histplot(x= 'fixed_acidity', data = df, kde= True, bins = 20)
```
